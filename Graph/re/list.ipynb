```python
class Node:
    def __init__(self, data):
        self.data = data  # The value/data held by the node
        self.next = None  # Pointer to the next node in the list

class LinkedList:
    def __init__(self):
        self.head = None  # Initialize the head of the list as None

    def is_empty(self):
        return self.head is None

    def append(self, data):
        """Append a new node with the given data to the end of the list."""
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return

        # Traverse to the end of the list
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def prepend(self, data):
        """Prepend a new node with the given data to the start of the list."""
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_after_node(self, prev_node_data, data):
        """Insert a new node with the given data after a node with specific data."""
        current = self.head
        # Find the node after which to insert
        while current and current.data != prev_node_data:
            current = current.next

        if current is None:
            print(f"Node with data {prev_node_data} not found.")
            return

        new_node = Node(data)
        new_node.next = current.next
        current.next = new_node

    def delete_node(self, key):
        """Delete the first node with the given key."""
        current = self.head
        previous = None

        # If the head node itself holds the key
        if current and current.data == key:
            self.head = current.next
            current = None
            return

        # Search for the key to be deleted, keep track of the previous node
        while current and current.data != key:
            previous = current
            current = current.next

        # If the key was not present in the list
        if current is None:
            print(f"Node with data {key} not found.")
            return

        # Unlink the node from the linked list
        previous.next = current.next
        current = None

    def display(self):
        """Print the contents of the list."""
        nodes = []
        current = self.head
        while current:
            nodes.append(str(current.data))
            current = current.next
        print(" -> ".join(nodes))

# Example usage:
ll = LinkedList()
ll.append(1)
ll.append(2)
ll.prepend(0)
ll.insert_after_node(1, 1.5)
ll.display()  # Output should display the list: 0 -> 1 -> 1.5 -> 2
ll.delete_node(1.5)
ll.display()  # Output should now be: 0 -> 1 -> 2
```

```python
class Node:
    def __init__(self, data):
        self.data = data  # Data stored in the node
        self.next = None  # Pointer to the next node

class LinkedList:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head is None

    def append(self, data):
        """Append a new node with the given data to the end of the list."""
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def prepend(self, data):
        """Prepend a new node with the given data to the start of the list."""
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_after_node(self, prev_node_data, data):
        """Insert a new node with the given data after the node containing prev_node_data."""
        current = self.head
        while current and current.data != prev_node_data:
            current = current.next

        if not current:
            print(f"Node with data {prev_node_data} not found.")
            return

        new_node = Node(data)
        new_node.next = current.next
        current.next = new_node

    def insert_at_index(self, index, data):
        """Insert a new node with the given data at the specified index.
        If index is 0, the node is prepended. Otherwise, traverse to the correct spot.
        """
        if index < 0:
            print("Negative index is not allowed.")
            return

        new_node = Node(data)

        if index == 0:
            # Insert at the head
            new_node.next = self.head
            self.head = new_node
            return

        current = self.head
        count = 0
        # Traverse to the node just before the insertion point
        while current is not None and count < index - 1:
            current = current.next
            count += 1

        if current is None:
            print(f"Index {index} is out of bounds.")
            return

        # Perform the insertion
        new_node.next = current.next
        current.next = new_node

    def delete_node(self, key):
        """Delete the first node that contains the given key."""
        current = self.head
        previous = None

        # If the head itself holds the key to be deleted
        if current and current.data == key:
            self.head = current.next
            current = None
            return

        # Search for the key to be deleted, track the previous node
        while current and current.data != key:
            previous = current
            current = current.next

        # If the key was not found
        if current is None:
            print(f"Node with data {key} not found.")
            return

        # Unlink the node from the linked list
        previous.next = current.next
        current = None

    def delete_at_index(self, index):
        """Delete the node at a specified index."""
        if index < 0:
            print("Negative index is not allowed.")
            return

        current = self.head
        previous = None
        count = 0

        # If the head needs to be removed
        if index == 0 and current:
            self.head = current.next
            current = None
            return

        # Traverse to the node at position index
        while current and count < index:
            previous = current
            current = current.next
            count += 1

        # If index is out of bounds
        if current is None:
            print(f"Index {index} is out of bounds.")
            return

        # Unlink the node from the list
        previous.next = current.next
        current = None

    def display(self):
        """Print the content of the list in order."""
        nodes = []
        current = self.head
        while current:
            nodes.append(str(current.data))
            current = current.next
        print(" -> ".join(nodes))

# Example usage:
ll = LinkedList()
ll.display() 

ll.append(1)
ll.display() 

ll.append(2)
ll.display() 

ll.prepend(0)
ll.display() 

ll.insert_after_node(1, 1.5)    # Insert 1.5 after the node containing 1
ll.display() 

ll.insert_at_index(2, 1.2)      # Insert 1.2 at index 2
ll.display()  # Expected: 0 -> 1 -> 1.2 -> 1.5 -> 2

ll.delete_at_index(2)           # Delete the node at index 2 (which has data 1.2)
ll.display()  # Expected: 0 -> 1 -> 1.5 -> 2

ll.delete_node(1.5)             # Delete node with data 1.5
ll.display()  # Expected: 0 -> 1 -> 2
```

```python
# Define a Node class to represent each element in the linked list
class Node:
    def __init__(self, data):
        self.data = data      # Store the node's data
        self.next = None      # Initialize pointer to the next node as None (no next node yet)

# Define a LinkedList class to manage and operate on the nodes
class LinkedList:
    def __init__(self):
        self.head = None      # Initialize the head of the list as None (empty list)

    def is_empty(self):
        # Check if the list is empty by verifying if head is None
        return self.head is None

    def append(self, data):
        # Create a new node with the provided data
        new_node = Node(data)
        # If the list is empty, set the new node as the head
        if not self.head:
            self.head = new_node
            return

        # Otherwise, find the last node in the list
        current = self.head
        while current.next:
            current = current.next  # Move to the next node until the end is reached
        # Set the next of the last node to the new node, appending it
        current.next = new_node

    def prepend(self, data):
        # Create a new node with the provided data
        new_node = Node(data)
        # Set the new node's next to the current head of the list
        new_node.next = self.head
        # Update the head to be the new node
        self.head = new_node

    def insert_after_node(self, prev_node_data, data):
        # Start at the head of the list
        current = self.head
        # Traverse until we find the node that contains the specified previous data
        while current and current.data != prev_node_data:
            current = current.next

        # If no such node was found, print a message and exit
        if not current:
            print(f"Node with data {prev_node_data} not found.")
            return

        # Create a new node with the provided data
        new_node = Node(data)
        # Link the new node's next to the current node's next, preserving the chain
        new_node.next = current.next
        # Link the current node to the new node, inserting it after current
        current.next = new_node

    def insert_before_node(self, next_node_data, data):
        """
        Insert a new node with the given data before the node that contains next_node_data.
        """
        # If the list is empty, there's nothing to insert before
        if not self.head:
            print("The list is empty.")
            return

        # Special case: if the head itself contains the data we want to insert before
        if self.head.data == next_node_data:
            # Use prepend to insert at the beginning of the list
            self.prepend(data)
            return

        prev = None             # Will hold the node before the target
        current = self.head     # Start from the head

        # Traverse until we find the node containing next_node_data
        while current and current.data != next_node_data:
            prev = current      # Keep track of the previous node
            current = current.next

        # If we've reached the end without finding next_node_data
        if current is None:
            print(f"Node with data {next_node_data} not found.")
            return

        # Create a new node with the provided data
        new_node = Node(data)
        # Set the previous node's next to the new node, linking before target
        prev.next = new_node
        # Link the new node's next to the current node which contains next_node_data
        new_node.next = current

    def insert_at_index(self, index, data):
        # Check for invalid negative index
        if index < 0:
            print("Negative index is not allowed.")
            return

        # Create a new node with the provided data
        new_node = Node(data)

        # Special case: insert at head if index is 0
        if index == 0:
            new_node.next = self.head  # New node points to current head
            self.head = new_node       # Update head to the new node
            return

        current = self.head  # Start traversal from head
        count = 0            # Counter to track current index
        # Traverse until we reach the node just before the desired index
        while current is not None and count < index - 1:
            current = current.next
            count += 1

        # If we've reached the end before reaching the desired index
        if current is None:
            print(f"Index {index} is out of bounds.")
            return

        # Insert the new node by adjusting pointers
        new_node.next = current.next  # New node points to node at insertion point
        current.next = new_node       # Previous node points to the new node

    def delete_node(self, key):
        current = self.head    # Start at the head
        previous = None        # To keep track of the previous node

        # Special case: if the head node itself needs to be deleted
        if current and current.data == key:
            self.head = current.next  # Move head to the next node
            current = None            # Clear the current node
            return

        # Traverse to find the node to delete
        while current and current.data != key:
            previous = current
            current = current.next

        # If the node was not found, print a message and exit
        if current is None:
            print(f"Node with data {key} not found.")
            return

        # Unlink the node from the list by linking previous node to current's next
        previous.next = current.next
        current = None   # Clear the node to be deleted

    def delete_at_index(self, index):
        # Check for invalid negative index
        if index < 0:
            print("Negative index is not allowed.")
            return

        current = self.head    # Start at the head of the list
        previous = None        # To keep track of the previous node
        count = 0              # Counter to track the current index

        # Special case: if we need to delete the head node
        if index == 0 and current:
            self.head = current.next  # Update head to next node
            current = None            # Clear the removed node
            return

        # Traverse to the node at the specified index
        while current and count < index:
            previous = current  # Save current as previous before moving on
            current = current.next
            count += 1          # Increment the index count

        # If we've reached the end without finding the index
        if current is None:
            print(f"Index {index} is out of bounds.")
            return

        # Unlink the node at the index from the list
        previous.next = current.next
        current = None  # Clear the removed node

    def display(self):
        nodes = []                # List to collect node data for display
        current = self.head       # Start from the head of the list
        # Traverse the entire list
        while current:
            nodes.append(str(current.data))  # Append current node's data as string
            current = current.next           # Move to next node
        # Print the linked list in a readable format separated by arrows
        print(" -> ".join(nodes))

# Example usage:
ll = LinkedList()  # Create an empty linked list
ll.display()       # Display the current list (should be empty)

ll.append(1)       # Append a node with data 1
ll.display()       # Display list: 1

ll.append(2)       # Append a node with data 2
ll.display()       # Display list: 1 -> 2

ll.append(4)       # Append a node with data 4
ll.display()       # Display list: 1 -> 2 -> 4

# Insert 1.5 before node with data 2
ll.insert_before_node(2, 1.5)
ll.display()       # Expected display: 1 -> 1.5 -> 2 -> 4

# Insert 0 before node with data 1 (which is at head)
ll.insert_before_node(1, 0)
ll.display()       # Expected display: 0 -> 1 -> 1.5 -> 2 -> 4
```

```python
# Define a Node class for the doubly linked list
class Node:
    def __init__(self, data):
        self.data = data      # Store the node's data
        self.prev = None      # Pointer to the previous node (None by default)
        self.next = None      # Pointer to the next node (None by default)

# Define the DoublyLinkedList class to manage the nodes
class DoublyLinkedList:
    def __init__(self):
        self.head = None      # Initialize the head of the list as None (empty list)

    def is_empty(self):
        # Return True if the list has no elements
        return self.head is None

    def append(self, data):
        # Create a new node with the provided data
        new_node = Node(data)
        # If the list is empty, set the new node as the head
        if not self.head:
            self.head = new_node
            return

        # Otherwise, traverse to the end of the list
        current = self.head
        while current.next:
            current = current.next  # Move to the next node
        # Link the new node at the end of the list
        current.next = new_node
        new_node.prev = current    # Set the previous pointer of the new node

    def prepend(self, data):
        # Create a new node with the provided data
        new_node = Node(data)
        # The new node's next pointer should point to the current head
        new_node.next = self.head
        # If the list isn't empty, update the current head's prev pointer
        if self.head:
            self.head.prev = new_node
        # Update the head of the list to be the new node
        self.head = new_node

    def insert_after_node(self, prev_node_data, data):
        # Start from the head of the list
        current = self.head
        # Traverse until we find the node with the specified data
        while current and current.data != prev_node_data:
            current = current.next

        # If the node isn't found, print an error message
        if not current:
            print(f"Node with data {prev_node_data} not found.")
            return

        # Create a new node with the provided data
        new_node = Node(data)
        # Set the new node's next pointer to the current node's next node
        new_node.next = current.next
        # Set the new node's previous pointer to the current node
        new_node.prev = current

        # If the current node isn't the last node, update the next node's prev pointer
        if current.next:
            current.next.prev = new_node

        # Insert the new node after the current node
        current.next = new_node

    def insert_before_node(self, next_node_data, data):
        """
        Insert a new node with the given data before the node containing next_node_data.
        """
        # If the list is empty, cannot insert before anything
        if not self.head:
            print("The list is empty.")
            return

        # Start from the head of the list
        current = self.head
        # Traverse until we find the target node containing next_node_data
        while current and current.data != next_node_data:
            current = current.next

        # If we didn't find such a node, print an error message
        if not current:
            print(f"Node with data {next_node_data} not found.")
            return

        # Create a new node with the provided data
        new_node = Node(data)
        # Set the new node's next pointer to the current (target) node
        new_node.next = current
        # Set the new node's previous pointer to the node before current
        new_node.prev = current.prev

        # If there is a node before current, update its next pointer to new_node
        if current.prev:
            current.prev.next = new_node
        else:
            # If current was the head, update the head to new_node
            self.head = new_node

        # Update the current node's prev pointer to the new node
        current.prev = new_node

    def insert_at_index(self, index, data):
        # Check for invalid negative index
        if index < 0:
            print("Negative index is not allowed.")
            return

        # Special case: inserting at the beginning
        if index == 0:
            self.prepend(data)
            return

        # Create a new node with the provided data
        new_node = Node(data)
        current = self.head   # Start traversal from head
        count = 0             # Index counter

        # Traverse until we reach the node just before the desired index
        while current is not None and count < index - 1:
            current = current.next
            count += 1

        # If we reached the end before the desired index
        if current is None:
            print(f"Index {index} is out of bounds.")
            return

        # Set the new node's next pointer to current's next
        new_node.next = current.next
        # Set the new node's prev pointer to current
        new_node.prev = current

        # If current is not the last node, update the next node's prev pointer
        if current.next:
            current.next.prev = new_node

        # Link current's next pointer to the new node, completing insertion
        current.next = new_node

    def delete_node(self, key):
        # Start from the head of the list
        current = self.head
        # Traverse until we find the node containing the key
        while current and current.data != key:
            current = current.next

        # If the node wasn't found, print an error message
        if not current:
            print(f"Node with data {key} not found.")
            return

        # If there's a node before the current one, update its next pointer
        if current.prev:
            current.prev.next = current.next
        else:
            # If current is the head, move the head pointer to next node
            self.head = current.next

        # If there's a node after the current one, update its prev pointer
        if current.next:
            current.next.prev = current.prev

        # Clear the current node
        current = None

    def delete_at_index(self, index):
        # Check for invalid negative index
        if index < 0:
            print("Negative index is not allowed.")
            return

        current = self.head  # Start at the head
        count = 0            # Index counter

        # Traverse until we reach the desired index
        while current and count < index:
            current = current.next
            count += 1

        # If we reached the end without finding the index
        if current is None:
            print(f"Index {index} is out of bounds.")
            return

        # If there's a previous node, update its next pointer
        if current.prev:
            current.prev.next = current.next
        else:
            # If current is the head, update head pointer
            self.head = current.next

        # If there's a next node, update its prev pointer
        if current.next:
            current.next.prev = current.prev

        # Clear the current node
        current = None

    def display(self):
        # Initialize an empty list to gather node data
        nodes = []
        current = self.head  # Start at the head of the list
        # Traverse through the list until the end
        while current:
            nodes.append(str(current.data))  # Append current node's data as a string
            current = current.next           # Move to the next node
        # Join all node data with a symbol to represent links and print the result
        print(" <-> ".join(nodes))

# Example usage:
dll = DoublyLinkedList()  # Create an empty doubly linked list
dll.display()             # Should display nothing as the list is empty

dll.append(1)             # Append a node with data=1
dll.display()             # Expected output: 1

dll.append(2)             # Append a node with data=2
dll.display()             # Expected output: 1 <-> 2

dll.prepend(0)            # Prepend a node with data=0 to the beginning
dll.display()             # Expected output: 0 <-> 1 <-> 2

dll.insert_after_node(1, 1.5)  # Insert 1.5 after node with data=1
dll.display()                  # Expected output: 0 <-> 1 <-> 1.5 <-> 2

dll.insert_before_node(2, 1.8) # Insert 1.8 before node with data=2
dll.display()                  # Expected output: 0 <-> 1 <-> 1.5 <-> 1.8 <-> 2

dll.delete_node(1.5)           # Delete the node with data=1.5
dll.display()                  # Expected output: 0 <-> 1 <-> 1.8 <-> 2

dll.delete_at_index(0)         # Delete the node at index 0 (data=0)
dll.display()                  # Expected output: 1 <-> 1.8 <-> 2
```

```python
class Stack:
    def __init__(self):
        # Initialize an empty list to store stack elements.
        # The end of the list will represent the top of the stack.
        self.items = []

    def is_empty(self):
        # Check if the stack is empty.
        # Returns True if there are no elements in the stack.
        return len(self.items) == 0

    def push(self, item):
        # Add an item to the top of the stack.
        # This is done by appending the item to the end of the list.
        self.items.append(item)

    def pop(self):
        # Remove and return the top element of the stack.
        # Before removing, check if the stack is empty.
        if self.is_empty():
            print("Stack is empty, cannot pop.")
            return None
        # Use list.pop() to remove and return the last element,
        # which represents the top of the stack.
        return self.items.pop()

    def peek(self):
        # Return the top element of the stack without removing it.
        # First, check if the stack is empty.
        if self.is_empty():
            print("Stack is empty, cannot peek.")
            return None
        # Access the last element of the list, representing the top.
        return self.items[-1]

    def size(self):
        # Return the number of elements in the stack.
        return len(self.items)

    def display(self):
        # Print the elements of the stack.
        # The bottom of the stack will be at the beginning of the list
        # and the top of the stack will be at the end of the list.
        print("Stack from bottom to top:", self.items)

# Example usage of the Stack class:
stack = Stack()          # Create a new empty stack
stack.display()          # Should show an empty list

stack.push(10)           # Push element 10 onto the stack
stack.push(20)           # Push element 20 onto the stack
stack.push(30)           # Push element 30 onto the stack
stack.display()          # Expected: Stack from bottom to top: [10, 20, 30]

print("Top element:", stack.peek())  # Expected: Top element: 30
print("Stack size:", stack.size())   # Expected: Stack size: 3

print("Popped element:", stack.pop())  # Expected: Popped element: 30
stack.display()                         # Expected: Stack from bottom to top: [10, 20]

print("Popped element:", stack.pop())  # Expected: Popped element: 20
print("Popped element:", stack.pop())  # Expected: Popped element: 10
print("Popped element:", stack.pop())  # Expected: Stack is empty, cannot pop. -> None

stack.display()  # Display final state of the stack (empty)
```

```python
class Queue:
    def __init__(self):
        # Initialize an empty list to store queue elements.
        # The beginning of the list represents the front of the queue,
        # and the end of the list represents the rear of the queue.
        self.items = []

    def is_empty(self):
        # Check if the queue is empty.
        # Returns True if there are no elements in the queue.
        return len(self.items) == 0

    def enqueue(self, item):
        # Add an item to the rear (end) of the queue.
        # This is done by appending the item to the end of the list.
        self.items.append(item)

    def dequeue(self):
        # Remove and return the front element of the queue.
        # Before removing, check if the queue is empty.
        if self.is_empty():
            print("Queue is empty, cannot dequeue.")
            return None
        # Remove and return the first element of the list,
        # which represents the front of the queue.
        return self.items.pop(0)

    def peek(self):
        # Return the front element of the queue without removing it.
        # First, check if the queue is empty.
        if self.is_empty():
            print("Queue is empty, nothing to peek.")
            return None
        # Return the first element of the list, representing the front of the queue.
        return self.items[0]

    def size(self):
        # Return the number of elements in the queue.
        return len(self.items)

    def display(self):
        # Print the elements of the queue.
        # The front of the queue will be at the beginning of the list
        # and the rear at the end of the list.
        print("Queue from front to rear:", self.items)

# Example usage:
queue = Queue()          # Create a new empty queue
queue.display()          # Should show an empty list: []

queue.enqueue(10)        # Enqueue element 10
queue.enqueue(20)        # Enqueue element 20
queue.enqueue(30)        # Enqueue element 30
queue.display()          # Expected: Queue from front to rear: [10, 20, 30]

print("Front element:", queue.peek())  # Expected: Front element: 10
print("Queue size:", queue.size())     # Expected: Queue size: 3

print("Dequeued element:", queue.dequeue())  # Expected: Dequeued element: 10
queue.display()                               # Expected: Queue from front to rear: [20, 30]
```
